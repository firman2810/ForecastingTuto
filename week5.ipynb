#IMPORT CSV

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/drive/MyDrive/mtcars.csv")
```

#BAR CHART

```python
# Group by cyl and calculate average mpg
avg_mpg = df.groupby("cyl")["mpg"].mean()

# Plot bar chart
plt.figure(figsize=(8,5))
ax = avg_mpg.plot(kind="bar",color="blue", edgecolor="black")

for i, value in enumerate(avg_mpg):
    ax.text(i, value, f"{value:.2f}", ha="center", va="bottom")

plt.xlabel("Number of cylinders")
plt.ylabel("Average miles per gallon (mpg)")
plt.title("Average miles per gallon by number of cylinders")
plt.xticks(rotation=0)

plt.tight_layout()
plt.show()
```

#HISTOGRAM

```python
plt.figure(figsize=(8,5))
plt.hist(df["hp"], bins=10, color="blue", edgecolor="black")

plt.title("Distribution of horsepower")
plt.xlabel("Horsepower")
plt.ylabel("Frequency")
plt.show()
```

#AREA CURVE

```python
df["mpg"].plot(kind="area", alpha=0.5, color ="green")
plt.title("Cumulatives MPG")
plt.xlabel("Cars")
plt.ylabel("Miles per gallon")
plt.show()
```

#PIE CHART

```python
# Count frequency of each gear type
gear_counts = df["gear"].value_counts()

# plot pie chart
plt.figure(figsize=(7,7))
plt.pie(gear_counts, labels=gear_counts.index, autopct="%1.1f%%", colors=["blue","green","red"], startangle=90)
plt.title("Gear Distribution by Cars")
plt.show()
```
